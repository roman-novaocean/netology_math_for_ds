```python
from sympy import diff, cos, sin
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

# Уровень 0

##### посчитайте значение 1-ой и 2-ой производной фунции
$$f(x) = x^5 + 4\cdot\sin(2x) + \cos(3x+3) $$
#### в точке x = 1

```python
# sympy

f1 = diff(x**5 + 4*sin(2*x) + cos(3*x+3))


print(f'1-я производная равна |    {f1}')
print(f'значение 1-ой производной в точке x = 1 |    {f1.subs(x,1)}')

f2 = diff(f1)

print(f'2-я производная равна |    {f2}')
print(f'значение 2-ой производной в точке x = 1 |    {f2.subs(x,1)}')
```

```python
# scipy

def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x+3)

f1_value = derivative(f, 1.0, dx=1e-6)
f2_value = derivative(f, 1.0, dx=1e-6, n = 2)

print(f'значение 1-ой производной в точке x = 1 |    {f1_value}')
print(f'значение 2-ой производной в точке x = 1 |    {f2_value}')
```

# Уровень 1

##### Постройте график первой и второй производной функции
$$f(x) = (\sin(2x+1))^5 $$
#### на отрезке [-5; 5]

```python
def f(x):
    return (sin(2*x+1))**5

x = np.linspace(-5, 5)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, f1x, label='derivative 1')
plt.plot(x, f2x, label='derivative 2')
plt.legend()
plt.show()
```
