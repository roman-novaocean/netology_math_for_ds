# Задание 1

```python
# простые числа до 15 это 2, 3, 5, 7, 11, 13
s = np.random.binomial(15, 0.5, 200000)
total = sum(s == 2) + sum(s == 3) + sum(s == 5) + sum(s == 7) + sum(s == 11) + sum(s == 13)
print(f'вероятность выпадения простого числа орлов при 15 подбрасываниях равна {total * 100 / 200000} %')
```

# Задание 2

```python
s = np.random.randint(15, size=(1000, 15))
unique_numbers = np.unique(s).size
print(f'число уникальных значений {unique_numbers}')

flat_s = s.flatten()

sns.distplot(flat_s, hist=True, kde=True, 
             bins=unique_numbers, color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 2})

print(f'среднее {np.mean(flat_s):.3}')
print(f'дисперсия {np.var(flat_s):.3}')
print(f'медиана {np.mean(flat_s):.3}')
```

##### Попарная независимость

$P(A1) = 1/2$

$P(A2) = 1/2$

$P(A3) = 1/2$

т.к. мы смотрим на 2 монеты, нас устраивают события 2 орла или 2 решки из 4 возможных


$P(A1 \cap A2)$ равна вероятности что все 3 монеты упали одной стороной, то есть 2 исхода(все орлы или все решки) из 8 возможных.

$P(A1 \cap A2) = 2/8 = 1/4 $ 

Что удовлетворяет условию попарной независимости:

$P(A1 \cap A2) = P(A1) * P(A2) = 1/2 * 1/2 = 1/4 $

То же самое для $P(A2 \cap A3)$  и  $P(A1 \cap A3)$

##### Независимость в совокупности

$P(A1 \cap A2 \cap A3) = 1/4$

вероятность этого события равна вероятности того, что все монеты упали одной стороной. Эта вероятность вычислена выше и равна 1/4 (2 исхода из 8 возможных). 

Но очевидно, что 

$P(A1 \cap A2 \cap A3) \neq P(A1) * P(A2) * P(A3)$ 

$ 1/4 \neq 1/2 * 1/2 * 1/2$

А значит, что события A1 A2 и A3 не независимы в совокупности. 

# Задание 4

A1 - 20% продукции из них 10% высшего сорта

A2 - 30% продукции из них 5% высшего сорта

A3 - 50% продукции из них 20% высшего сорта

Формула полной вероятности:
$P(A) = \sum_{i = 1}^{n} P(A|B_{i}) P(B_{i})$

P(A1) = 0.1 * 0.2 = 0.02

P(A2) = 0.05 * 0.3 = 0.015

P(A3) = 0.2 * 0.5  = 0.1

P(A) = P(A1) + P(A2) + P(A3) = 0.135

Вероятность того, что случайно купленная продукция высшего сорта равна 13.5%
