```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
```

# Задание 1

##### Normal Distribution

```python
mu1 = 2.0
mu2 = 1.0
mu3 = 3.0
sigma1 = 0.5
sigma2 = 1.5
sigma3 = 1.0

norm_rv1 = sts.norm(loc=mu1, scale=sigma1)
norm_rv2 = sts.norm(loc=mu2, scale=sigma1)
norm_rv3 = sts.norm(loc=mu3, scale=sigma1)
norm_rv4 = sts.norm(loc=mu1, scale=sigma2)
norm_rv5 = sts.norm(loc=mu1, scale=sigma3)
```

```python
x = np.linspace(-2,6,100)
pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf1, label="mu=2.0 sigma=0.5")
plt.plot(x, pdf2, label="mu=1.0 sigma=0.5")
plt.plot(x, pdf3, label="mu=3.0 sigma=0.5")
plt.plot(x, pdf4, label="mu=2.0 sigma=1.5")
plt.plot(x, pdf5, label="mu=2.0 sigma=1.0")

plt.ylabel('$P(x)$')
plt.xlabel('$x$')
plt.legend(bbox_to_anchor=(1.5, 0.5), loc = 'center right')
plt.show()
```

##### Exponential Distribution

```python
lam1 = 0.1
lam2 = 0.5
lam3 = 1.0
lam4 = 2.0
lam5 = 4.0

expon_rv1 = sts.expon(lam1)
expon_rv2 = sts.expon(lam2)
expon_rv3 = sts.expon(lam3)
expon_rv4 = sts.expon(lam4)
expon_rv5 = sts.expon(lam5)
```

```python
x = np.linspace(-1,10,100)
pdf1 = expon_rv1.pdf(x)
pdf2 = expon_rv2.pdf(x)
pdf3 = expon_rv3.pdf(x)
pdf4 = expon_rv4.pdf(x)
pdf5 = expon_rv5.pdf(x)

plt.plot(x, pdf1, label="lamda=0.1")
plt.plot(x, pdf2, label="lamda=0.5")
plt.plot(x, pdf3, label="lamda=1.0")
plt.plot(x, pdf4, label="lamda=2.0")
plt.plot(x, pdf5, label="lamda=4.0")


plt.ylabel('$P(x)$')
plt.xlabel('$x$')
plt.legend()
plt.show()
```

##### t Distribution

```python
n1 = 1
n2 = 2
n3 = 4
n4 = 8
n5 = 100

t_rv1 = sts.t(n1)
t_rv2 = sts.t(n2)
t_rv3 = sts.t(n3)
t_rv4 = sts.t(n4)
t_rv5 = sts.t(n5)
```

```python
x = np.linspace(-5,5,100)
pdf1 = t_rv1.pdf(x)
pdf2 = t_rv2.pdf(x)
pdf3 = t_rv3.pdf(x)
pdf4 = t_rv4.pdf(x)
pdf5 = t_rv5.pdf(x)

plt.plot(x, pdf1, label="df=1")
plt.plot(x, pdf2, label="df=2")
plt.plot(x, pdf3, label="df=4")
plt.plot(x, pdf4, label="df=8")
plt.plot(x, pdf5, label="df=100")


plt.ylabel('$P(x)$')
plt.xlabel('$x$')
plt.legend()
plt.show()
```

# Задание 3

Энтропия Шэннона

$ {\displaystyle \mathrm {H} (X)=-\sum _{i=1}^{n}{\mathrm {P} (x_{i})\log \mathrm {P} (x_{i})}} $

```python
p_red = 54/86
p_yellow = 4/86
p_green = 28/86

H = - (p_red * np.log2(p_red) + p_yellow * np.log2(p_yellow) + p_green * np.log2(p_green))
print('количество информации которое несёт сообщение о цвете светофора в текущий момент =', H)
```
